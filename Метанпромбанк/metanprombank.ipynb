```python
#Создадим дополнительный датафрейм для дальнейшей корреляции по всем параметрам, 
# userid сделаем индексом, а качественные показатели закодируем.
df = data.copy()
df['balance_cod'] = df['balance'].apply(lambda x: 1 if pd.isna(x) else 0)
df = df.drop('balance', axis =1)
df.head()
```

```python
phik_overview = df.phik_matrix(interval_cols= ['score','est_salary','age']).round(2)
phik_overview
```

```python
plt.figure(figsize=(7, 10))
sns.heatmap(phik_overview.loc[:,['balance_cod']].drop('balance_cod').sort_values('balance_cod'), annot=True, cmap='coolwarm',center=0, linecolor='black', linewidths=1)
plt.title(f'Матрица корреляций')
plt.show()
```

Согласно шкале Чеддока:  
У пропусков в Balance существет корреляция:  
**Весьма высокая**  с equity,  
**Высокая** score,  
**Заметная** userid,  
**Умеренная** product, est_salary и churn  
**Слабая** credit_card и gender. Следовательно с такой зависимостью исключать пропуски не целесообразно.

По Анализу проведенному выше пропуски в столбцe `age` являются MCAR (Missing Completely At Random / Отсутствует совершенно случайно). Заполнять данные пропуски медианной не имеет смысла. <br>
А пропуске в `balance` - MNAR (Missing Not At Random / Отсутствует не случайно) — пропуски зависят от данных во многих столбцах, без дополнительного обоснования их нельзя отбрасывать или заполнять одним значением, т.к. это приведёт к заметным искажениям.

```python
#Для упрощения будем использовать функцию расчета долей параметров

def param_perc(column):
    param = []
    param = data.pivot_table (index = column, values = 'churn', aggfunc = ['sum', 'count'])
    param.columns = param.columns.droplevel(level=1)
    param['perc'] = round(param['sum']/param['count'],2)
    return param
```

## Проверка статистических гипотез  
**Гипотеза №1:**   
Проверим гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались.
  
Гипотезы:<br>
H0: доходы оставшихся и отточных клиентов равны;<br>
H1: доходы оставшихся и отточных клиентов не равны.<br>

Для определения используемого Т-теста проверим на три условия:

- Генеральные совокупности не зависят друг от друга.
- Выборочные средние нормально распределены.
- Дисперсии рассматриваемых генеральных совокупностей равны.

```python
alpha = 0.05

shapiro_test_stat, p_value1 = st.shapiro(data[data['churn'] == 1]['est_salary'])

print('p-значение: ', p_value1)
if p_value1 < alpha:
    print('Распределение не нормально для ушедших клиентов')
else:
    print('Нормальное распределение для ушедших клиентов')
shapiro_test_stat, p_value0 = st.shapiro(data[data['churn'] == 0]['est_salary'])
print('p-значение: ', p_value0)
if p_value0 < alpha:
    print('Распределение не нормально для оставшихся клиентов')
else:
    print('Нормальное распределение для оставшихся клиентов')       
```

```python
# оценим дисперсии
print('Диспертия ушедших клиентов',data[data['churn'] == 1]['est_salary'].var())
print('Диспертия оставшихся клиентов',data[data['churn'] == 0]['est_salary'].var())
```

```python
plt.figure(figsize=(15, 5))
sns.distplot(data.query('churn== 0')['est_salary'], label='остались')
sns.distplot(data.query('churn== 1')['est_salary'], label='ушли')
plt.ylabel('клиенты')
plt.xlabel('est_salary')
plt.grid()
plt.title('Распределение доходов действующих и ушедших клиентов')
plt.legend()
plt.show();
```

Выглядит ненормальным.

Генеральные совокупности не зависят друг от друга;<br>
Распределение значений не является нормальным;<br>
Дисперсии не равны.<br>
Можно использовать критерий Манна - Уитни `.mannwhitneyu()`<br>
Но ttest более интерпретируемый и понятный для бизнеса, и его можно использовать используюя параметр `equal_var`

```python
alpha = 0.05
results = st.ttest_ind(data[data['churn']==0]['est_salary'], 
                       data[data['churn']==1]['est_salary'],
                       equal_var=False)
print('p-значение: ', results.pvalue)    # тест двухсторонний

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Cчитаем, что доходы оставшихся и отточных клиентов не имеют разницы.
