### Исправим даты:

```python
def to_date(column): # для упрощения создадим функцию.
    data[column] = pd.to_datetime(data[column], unit='s').astype('datetime64[s]') # так же поменяем тип даты с object на datetime64
    data[column] = data[column].dt.round('1s') # округляем дату до секунд
```

```python
to_date('ts')
to_date('latest_reply')
to_date('thread_ts')
```

Поменяем тип у столбца file_date

```python
data['file_date']=data['file_date'].astype('datetime64[s]')
display(data.head())
```

Теперь можно определить временной диапазон полученных данных.

```python
def renge_time (times):
    min_time = data[times].min()
    max_time = data[times].max()
    print(f'начальное время полученных данных -', min_time)
    print(f'последнее время полученных данных -', max_time)
    print(f'общее время -', max_time - min_time)
```

```python
renge_time('ts')
```

Рассмотрим временные диапазоны колонок `thread_ts` и `file_date` и их распределение по диапазону `ts`

```python
renge_time('thread_ts')
renge_time('file_date')
```

Оказалось, что диапазон даты тредов в два раза меньше, чем диапазон даты постов и охватывают только половину рассмотренного периода.

Столбец `reactions` выглядит не информативно, разобьем его на 3: names, users, counts и посчитаем общее число реакций

```python
# Функция для разбора столбца 'reactions'
def drop_reactions(reaction_str):
    try:
        return ast.literal_eval(reaction_str)
    except (ValueError, SyntaxError):
        return []
```

```python
def jr_replace(x):
    match = re.sub(r'_1$|_2$|_3$|_blue$|_red$|_yellow$|_green$|_01$|_02$|_03$|_orange$|_purple$|_new$|_me$',"",x)
    return match
data['channel'] = data['channel'].apply(jr_replace)
len(data['channel'].unique())
print('Всего уникальных: ', len(data['channel'].unique()))
```

```python
data['cohorts']=data['channel']
```

```python
def cohort (df: pd.DataFrame, col: str):
    for i in range(len(df[col])):
        df[col][i] = re.split("_|-",df[col][i])
        a =df[col][i]
        try:
            if df[col][i][0]=='data' :
                df[col][i] = a[-2]
            elif df[col][i][0]=='masterskaya':
                df[col][i] = a[0]
            else:
                df[col][i] = a[-1]
        except:
            df[col][i] = df[col][i]

cohort(data,'cohorts')
data['cohorts']=data['cohorts']
for i in range(len(data['cohorts'])):
    data['cohorts'][i] = ''.join(data['cohorts'][i])
```

## Исследовательский анализ данных

Создадим функцию для построения графиков.

```python
def plot_mes(x,y, title, xlable, ylable): # описание функции и их переменные. Х-значения по оси Х, Y- значения по оси Y
    unit = data.loc[data[x].notna(), [y, x]] \
        .groupby(y) \
        .agg(['count']) 
    unit.columns = unit.columns.get_level_values(1)
    ax = unit.plot.bar(figsize=(12,6)) 
    plt.title(title)
    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.show()
```

### Сообщения.

Используем функцию для построения однотипных графиков.

```python
def plot_rat_time(x,y, ttle):
    data[[x,y]] \
    .groupby(y).agg('sum') \
    .plot(title=ttle, kind='bar')

def plot_rat(x,y, ttle):
    data[[x,y]] \
    .groupby(y).agg('count') \
    .sort_values(by=x, ascending=False) \
    .plot(title=ttle, kind='bar')
```

```python
plot_rat_time('r_counts','ts_h','Количество реакций в зависимости от часа')
plot_rat_time('r_counts','ts_w','Количество реакций в зависимости от дня недели')
plot_rat_time('r_counts','ts_mnth','Количество реакций в зависимости от месяца')
plot_rat_time('r_counts','ts_y','Количество реакций в зависимости от года')
plot_rat('r_counts','cohorts','Количество реакций в зависимости от канала')
plot_rat('r_counts','profession','Количество реакций в зависимости от профессий')
```

Создадим функцию для однотипных графиков

```python
def plot_usrs(x,y,ttle,xlable,ylable,agg,x2,y2,ttle2,xlable2,ylable2,agg2):
    a = data[[x,y]] \
        .groupby(y).agg(agg) \
        .sort_values(by=x, ascending=False)
    a = a.head(30)
    ax1 = plt.subplot(1, 2, 1)
    a.plot.bar(figsize=(12,6),grid=True,title = ttle,ax=ax1)
    
    b = data[[x2,y2]] \
        .groupby(y2).agg(agg2) \
        .sort_values(by=x2, ascending=False)
    b = b.head(30)
    ax2 = plt.subplot(1, 2, 2,sharey=ax1)
    b.plot.bar(figsize=(12,6),grid=True,title = ttle2,ax=ax2)
```

### Приток и отток пользователей

Создадим функцию для однотипных графиков.

```python
def plt_subtipe (x,y,title_1,title_2,status_1, status_2,xlabel):
    a = data.loc[data['subtype'] == status_1]
    a = a.groupby(x) \
         .agg({y:'count'})
    b = data.loc[data['subtype'] == status_2]
    b = b.groupby(x) \
         .agg({y:'count'})
# первый график
    ax1 = plt.subplot(1, 2, 1)
    a.plot.bar(figsize=(12,6),grid=True, ax=ax1)
    plt.legend()
    plt.xlabel(xlabel)
    plt.title(title_1)
# второй график
    ax2 = plt.subplot(1, 2, 2, sharey=ax1)
    b.plot.bar(figsize=(12,6),grid=True, ax=ax2)
    plt.xlabel(xlabel)
    plt.title(title_2)
```

```python
plt_subtipe('ts_h','user','кол-во пользователей присоединилось (по часам)','кол-во пользователей покинуло (по часам)','channel_join','channel_leave','часы')
```

```python
plt_subtipe('ts_mnth','user','кол-во пользователей присоединилось(по месяцам)','кол-во пользователей покинуло (по месяцам)','channel_join','channel_leave','месяц')
```

Меньше всего пришло пользователей по времени с 20:00 до 4:00, по дням недели - в воскресенье, по месяцам в январе.  
  
Больше покинуло пользователей чаты по часам в 9:00 и 14:00, по дням недели - в понедельник, по месяцам в декабре.  
Меньше всего пришло пользователей по времени с 20:00 до 4:00, по дням недели - в субботу, по месяцам в январе.   
  
Покидает число пользователей больше, чем присоединяются новые. Это явный отток пользователей.  
В основном действия совершаются в один и тот же промежуток времени. Данной динамикой можно воспользоваться для удержания пользователей.

```python
# график пришедших в канал пользователей
ax1 = data.query('subtype == "channel_join"') \
        .groupby('cohorts') \
        .agg(join =('user', 'count')) \
        .plot(kind='bar', figsize=(10, 5), color='darkblue', alpha=0.4)

# график покинувших в канал пользователей
ax2 = data.query('subtype == "channel_leave"') \
        .groupby('cohorts').agg(leave =('user', 'count')) \
        .plot(kind='bar',figsize=(10, 5),color='lime',alpha=0.4,ax=ax1)
plt.title('Сравнение пришедших и покинутых пользователей в каналах', fontsize=15);
```
